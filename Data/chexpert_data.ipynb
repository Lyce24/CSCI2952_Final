```python
import pandas as pd

# Downloaded from https://aimi.stanford.edu/datasets/chexpert-chest-x-rays
df = pd.read_csv("../data/train_visualCheXbert.csv")

# Process the Path column: keep parts from index 2 onward
df["Path"] = df["Path"].apply(lambda x: "/".join(x.split("/")[2:]))

df.to_csv("train_split.csv", index=False)
```

```python
# ---- Counts for each column ----
print("=== Value Counts per Column ===\n")
index = 0
for col in df.columns:
    if index==0:
        index += 1
        continue
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False))  # keep NaNs visible

# ---- Summary statistics (mean, std, min, max, etc.) ----
print("\n\n=== Numeric Summary (mean, std, etc.) ===\n")
print(df.describe(include=[float, int]).T[['mean', 'std', 'min', 'max']])

# ---- Optional: summarize categorical columns (unique, top, freq) ----
print("\n\n=== Categorical Summary ===\n")
print(df.describe(include=['object', 'category']).T[['unique', 'top', 'freq']])
```
