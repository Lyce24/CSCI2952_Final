```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("raddar/tuberculosis-chest-xrays-shenzhen")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv("../../../.cache/kagglehub/datasets/raddar/tuberculosis-chest-xrays-shenzhen/versions/1/shenzhen_metadata.csv")

# if findings == "normal", label 0, else label 1
df['Label'] = df['findings'].apply(lambda x: 0 if x == "normal" else 1)

# print distribution of labels
print(df['Label'].value_counts())
```

```python
# new df with only filename and label columns
# rename study_id to Path
df_tb = df[['study_id', 'Label']].rename(columns={'study_id': 'Path'})

df_tb
```

```python
from sklearn.model_selection import train_test_split

# test size 20%, val size 12%, train size 68%

train_df, test_df = train_test_split(df_tb, test_size=0.2, stratify=df_tb['Label'], random_state=42)
train_df, val_df = train_test_split(train_df, test_size=0.15, stratify=train_df['Label'], random_state=42)

print("Train size:", len(train_df))
print("Val size:", len(val_df))
print("Test size:", len(test_df))

train_df.to_csv("../src/tb_train_split.csv", index=False)
val_df.to_csv("../src/tb_val_split.csv", index=False)
test_df.to_csv("../src/tb_test_split.csv", index=False)
```
