```python
import pandas as pd

# create a dataframe from data_dict
data_list = []
for file_name, (label, split) in data_dict.items():
    data_list.append({"Path": file_name, "Label": label, "Split": split})

df = pd.DataFrame(data_list)
df.head()
```

```python
# in the dataframe, replace label "COVID-19" with 1, Non-COVID with 2, Normal with 0
df["Label"] = df["Label"].replace({"COVID-19": 1, "Non-COVID": 2, "Normal": 0})
df.head()
```

```python
# create train, val, test splits
train_df = df[df["Split"] == "Train"].reset_index(drop=True)
val_df = df[df["Split"] == "Val"].reset_index(drop=True)
test_df = df[df["Split"] == "Test"].reset_index(drop=True)

print("Train size:", len(train_df))
print("Val size:", len(val_df))
print("Test size:", len(test_df))

train_df.to_csv("../src/covidqu_train_split.csv", index=False)
val_df.to_csv("../src/covidqu_val_split.csv", index=False)
test_df.to_csv("../src/covidqu_test_split.csv", index=False)
```

```python
# for each dataframe, print the distribution of labels
print("Train label distribution:")
print(train_df["Label"].value_counts())
print("Val label distribution:")
print(val_df["Label"].value_counts())
print("Test label distribution:")
print(test_df["Label"].value_counts())
```
