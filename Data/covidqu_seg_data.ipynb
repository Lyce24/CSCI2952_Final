```python
import os
import pandas as pd


BASE = "/users/kmaeda2/scratch/QUEx/Lung Segmentation Data/Lung Segmentation Data"


splits = ["Train", "Val", "Test"]


rows = {s: [] for s in splits}


for split in splits:
   split_dir = os.path.join(BASE, split)


   for category in ["COVID-19", "Non-COVID", "Normal"]:
       img_dir = os.path.join(split_dir, category, "images")
       mask_dir = os.path.join(split_dir, category, "lung masks")


       for fname in sorted(os.listdir(img_dir)):
           if not fname.lower().endswith(".png"):
               continue


           img_rel = f"{split}/{category}/images/{fname}"


           mask_rel = f"{split}/{category}/lung masks/{fname}"


           if not os.path.exists(os.path.join(BASE, mask_rel)):
               continue


           rows[split].append([img_rel, mask_rel])


train_df = pd.DataFrame(rows["Train"], columns=["img_path", "mask_path"])
val_df = pd.DataFrame(rows["Val"], columns=["img_path", "mask_path"])
test_df = pd.DataFrame(rows["Test"], columns=["img_path", "mask_path"])
```

```python
out_dir = os.path.join(BASE, "splits")
os.makedirs(out_dir, exist_ok=True)


train_df.to_csv(os.path.join(out_dir,"train.csv"), index=False)
val_df.to_csv(os.path.join(out_dir,"val.csv"), index=False)
test_df.to_csv(os.path.join(out_dir,"test.csv"), index=False)


print("Saved:")
print(os.path.join(out_dir, "train.csv"))
print(os.path.join(out_dir, "val.csv"))
print(os.path.join(out_dir, "test.csv"))
```
