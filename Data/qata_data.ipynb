```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

BASE = "/users/kmaeda2/scratch/QaTa/QaTa-COV19/QaTa-COV19-v2"

train_img_dir = os.path.join(BASE, "Train Set", "Images")
train_mask_dir = os.path.join(BASE, "Train Set", "Ground-truths")

test_img_dir = os.path.join(BASE, "Test Set", "Images")
test_mask_dir = os.path.join(BASE, "Test Set", "Ground-truths")

rows_train = []
rows_test = []

# train
for img in sorted(os.listdir(train_img_dir)):
    if not img.endswith(".png"):
        continue
    mask = f"mask_{img}"
    mask_path = os.path.join(train_mask_dir, mask)

    if not os.path.exists(mask_path):
        continue

    rows_train.append([
        f"Train Set/Images/{img}",
        f"Train Set/Ground-truths/{mask}"
    ])

df_train = pd.DataFrame(rows_train, columns=["img_path", "mask_path"])
train_df, val_df  = train_test_split(df_train, test_size=0.1, random_state=42, shuffle=True)

# test
for img in sorted(os.listdir(test_img_dir)):
    if not img.endswith(".png"):
        continue
    mask = f"mask_{img}"
    mask_path = os.path.join(test_mask_dir, mask)

    if not os.path.exists(mask_path):
        continue

    rows_test.append([
        f"Test Set/Images/{img}",
        f"Test Set/Ground-truths/{mask}"
    ])

df_test = pd.DataFrame(rows_test, columns=["img_path", "mask_path"])


out_dir = os.path.join(BASE, "splits")
os.makedirs(out_dir, exist_ok=True)

train_df.to_csv(os.path.join(out_dir, "train.csv"), index=False)
val_df.to_csv(os.path.join(out_dir, "val.csv"), index=False)
df_test.to_csv(os.path.join(out_dir, "test.csv"), index=False)

print("Saved:")
print(os.path.join(out_dir, "train.csv"))
print(os.path.join(out_dir, "val.csv"))
print(os.path.join(out_dir, "test.csv"))
```
